**Kiểu dữ liệu**

- Dữ liệu định lương (Quantitative data) : là dữ liệu có thể do đếm được (rời rạc , liên tục)

- Dữ liệu định tính (Qualitative data) : là dữ liệu sử dụng để phân loại , giá trị của dữ liệu được sử dụng để đại diện cho một phân loại nào đó (có thứ tự , không có thứ tự)

**Thu thập dữ liệu mẫu**

- Từ quan sát (observational study) : là dữ liệu có được bằng cách quan sát và đo lường được mà không tác động hay tìm cách thay đổi đối tượng nghiên cứu

- Từ thử nghiệm (experiment): là dữ liệu được bằng cách áp dụng một số phương pháp tác động lên đối tượng cần nghiên cứu sau đó quan sát ảnh hưởng của phương pháp lên đối tượng đó

- Simple random sample : là cách lấy mẫu mà mỗi giá trị dữ liệu được lấy từ quần thể theo cùng một cách và cơ hội được chọn của mỗi giá trị dữ liệu là như nhau

- Systematic sampling : là lấy mẫu bằng cách chọn một điểm bắt dầu và điểm kết thúc , sau đó lần lượt chọn phần tử thứ k trong quần thể

- Convenience sampling : là cách lấy mẫu mà kết quả được thu thập một cách dễ dàng

- Stratified sampling : chia quần thể thành nhiều nhóm nhỏ , mỗi nhóm có cùng đặc tính , sau đó lấy mẫu bằng cách chọn ngẫu nhiên từ các nhóm nhỏ đó

- Cluser Sampling : chia quần thể thành nhiều cụm nhỏ , chọn ngẫu nhiên một số cụm và lấy tất cả các dữi liệu của các cụm được chọn

- Multistage sampling : lấy nhiều giai đoạn : là phương pháp lấy mẫu bằng cách kết hợp nhiều phương pháp lấy mẫu đơn giản với nhau

**Đặc tính của dữ liệu**

- Central tendency : thể hiện vị trí mà phần lớn tập dữ liệu tập trung

- Variation : thể hiện sự phân tán của các giá trị dữ liệu

- Phân bố : hình dạng của dữ liệu khi sắp xếp theo giá trị

- Outliers : các giá trị nằm cách xa so với hầu hết các giá trị khác trong tập dữ liệu

```python
# Đồ thị Stem & Leaf

data = [2 , 12,15,17,21,23,25]

dct = {}

for x in data :
  d = x % 10
  x //= 10
  if x not in dct :
    dct[x] = [d]
  else :
    dct[x].append(d)


for key , values in dct.items() :
  print(f'{key} : {values}')
```

# **Descriptive statistics**

Mô tả những đặc tính quan trọng của tập dữ liệu mẫu bằng các số liệu thống kê như giá trị trung bình , độ lệch chuẩn ... (Từ population -> sample)



## **Central tendency** :

- Mean :
$\bar{x}$ : giá trị trung bình của sample

$$
\bar{x} = \frac{\sum_{i = 0}^{n}x_{i}}{n}
$$

$\mu$ giá trị trung bình của tất cả các giá trị của quần thể

$$
\mu = \frac{\sum_{i = 0}^{n}x_{i}}{N}
$$

- Ưu điểm :

  + Tất cả các giá trị của dữ liệu điều được dùng để tính trung bình
  + Trung bình mẫu có xu hướng chênh lệch không nhiều so với trung bình quần thể

- Khuyết diểm :

  + Nhạy cảm với outliers  

- Trung bình trọng số :

$$
\bar{x} = \frac{\sum(w.x)}{\sum w}
$$

- Median : là giá trị nằm giữa khi tập dữ liệu được sắp xếp ; $\tilde{x}$ ; không bị ảnh hưởng bởi các giá trị ngoại lệ ; Lẻ : ở giữa ; Chắn : $\frac{x_{l} + x_{r}}{2}$

- Mode : là giá trị xuất hiện nhiều lần nhất ; một tập có một hoặc nhiều yếu vị , hoặc có thể không có ; chỉ được dùng làm độ đo xu hướng hướng tâm đối với dữ liệu không có thứ tự



## **Variability** :

  ~ Range : độ lệch giữa giá trị lớn nhất và giá trị bé nhất của dữ liệu ; là độ đo biến thiên đơn giản nhất ; bị ảnh hưởng bởi outliers

  ~ Standard deviation ($s$): đại lượng chỉ độ lệch của các giá trị dữ liệu so với giá trị trung bình   

$s$ :  độ lệch chuẩn của mẫu:

$$
s = \sqrt{\frac{\sum(x - \bar{x})^{2}}{n - 1}}
= \sqrt{\frac{n(\sum x^{2}) - (\sum x)^{2}}{n(n - 1)}}
$$

Giá trị độ lệch chuẩn không âm ; bị ảnh hưởng bởi outliers ; đơn vị của std cũng là đơn vị của giá trị dữ liệu

$\sigma$ :  độ lệch chuẩn của quần thể :

$$
\sigma = \sqrt{\frac{\sum(x - \mu)^{2}}{N}}
$$

## **The Empirical Rule**

- Nếu tập dữ liệu có dạng hình chuông hoặc xấp xỉ dạng hình chuông

  ~ Khoảng $68\%$ dữ liệu trong đoạn [$\mu - \sigma , \mu + \sigma$]

  ~ Khoảng $95\%$ dữ liệu nằm trong [$\mu - 2\sigma , \mu + 2\sigma$]

  ~ Khoảng $99.7\%$ dữ liệu nằm trong đoạn [$\mu-3\sigma , \mu + 3\sigma$]

## **Định lý Chebyshev** :
- Với một tổng thể bất kỳ có trung bình $\mu$ và std $\sigma$

- Phần trăm các giá trị quan sát nằm trong khoảng [$\mu - k\sigma ; \mu + k\sigma$] bằng ít nhất (1 - $\frac{1}{k^{2}}$) với $k > 1$

## **Coeficient of variation** (hệ số biến thiên) :
được sử dụng để so sánh sự biến thiên của hai hay nhiều tập dữ liệu và mô tả mối liên hệ giữa độ lệch chuẩn và trung bình ; \% ; Công thức

- Công thức :

  ~ Sample : $cv = \frac{s}{\bar{x}}.100\%$

  ~ Population : $cv = \frac{\sigma}{\mu}.100\%$

## **Z-score**

- Là sự khác biệt của giá trị dữ liệu với giá trị trung bình của tập dữ liệu được đo bằng số lần của độ lệch chuẩn ; tròn với sai số là 0.01

- Sample : $z = \frac{x - \bar{x}}{s}$

- Population : $z = \frac{x - \mu}{\sigma}$

## **Gauss distribution** :

Xét trong miền ($-\infty , +\infty$) , biến ngẫu nhiên $X$ có hàm mất độ xác suất PDF ứng với hai tham số thực $\mu , \sigma^2$ là :
$$
f_{X}(x) = \frac{1}{\sigma \sqrt{2\pi}} . e^{-\frac{1}{2\sigma^{2}}(x - \mu)^2} , \forall x \in R
$$

được gọi là phân phối chuẩn .

## **Phân phối Z**

là phân phối chuẩn chính tắc : $\mu$ , $\sigma^{2} = 1$

$$
f(z) = \frac{1}{\sqrt{2\pi}} .e^{-\frac{z^2}{2}}
$$

- Critial value là giá trị $z$ mà tại đó phần diện tích bên phải $z$ bằng một giá trị $\alpha$ :

nếu $\alpha = 0.025$ thì critical value là $z_{0.025} = 1.96$ , nghĩa là $z_{0.025} = 1.96$ có diện tích phần bên phải là $0.025$

- Độ tin cậy $= 1 - \alpha$



## **Sampling distribution** :
Với population và cỡ mẫu n , sample dítribution của 1 giá trị thông kê là phân phối tất cả các giá trị thống kê cho tất cả các mẫu có thể có với kích thước $n$

- Xét sampling distribution của trung bình là phân phối của tất cả các giá trị trung bình mẫu (sample mean) nếu như ta liệt kê được mọi tập mẫu có thể lấy với kích thước cố định

  ~ Giá trị trung bình của trung bình mẫu : $\mu_{\bar{x}}$

  ~ Độ lệch chuẩn của trung bình mẫu : $\sigma_{\bar{x}}$

```python
# Ex : sampling distribution
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

population = np.array([18 , 20 , 22 , 24])
mu = population.mean()
sigma = population.std().round(3)
N = len(population)

print(mu , sigma , N )

x_ = []
for i in population :
  for j in population :
    val = (i+j) / 2
    x_.append(val.astype(int))

x_ = np.array(x_)
print(x_)

plt.hist(x_ , bins=range(min(x_), max(x_)+2) )

plt.xlabel('Sample Mean')
plt.ylabel('Freq')
plt.title('Sampling distribution of mean with n = 2')

plt.show()
```

## **Định lý Giới hạn trung tâm**

  ~ Giá trị trung bình của phân phối mẫu : $\mu_{\bar{x}} = \mu$

  ~ Phương sai của phân phối mẫu : $\sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}}$



## **Bộ ước lượng lệch(biased) và không lệch(unbiased)**

- Bộ ước lượng không lệch : Một thống kê có kỳ vọng (giá trị trung bình của phân phối mẫu) bằng tham số của tổng thể

- Trung bình mẫu : không lệch với trung bình của quần thể

- Tỉ lệ mẫu : không lệch với tỉ lệ của quần thể

- Phương sai mẫu : không lệch với phương sai của quần thể

- Sample range : lệch so với range của quần thể

## **Estimation** :

- **Point estimation** : xác định một giá trị số là giá trị ước lượng cho tham số quần thể

- **Interval estimation** : xác định một khoảng giá trị có nhiều khả năng chứa giá trị tham số quần thể


  + EX : Biến ngẫu nhiên $X$ phân phối chuẩn với kỳ vọng $\mu$ chưa biết . Trung bình mẫu $\bar{x}$ là một bộ ước lượng điểm cho kỳ vọng $\mu$ ; [25 , 30 , 29 , 31] :

  + Giá trị ước lượng của điểm $\mu$ :

  $$\bar{x} = \frac{25 + 30 + 29 + 31}{4} = 28.75$$
  
  + Kỳ vọng $\sim 28.75$  ; Kỳ vọng $\in [28 , 29.5]$

### **Ước lượng trung bình quần thể**

1. Khái niệm chính

- Giá trị trung bình $\bar{x}$ của mẫu là ước lượng điểm tốt nhất của trung bình quần thể $\mu$

- Sử dụng dữ liệu mẫu để xây dựng một khoảng ước lượng cho giá trị trung bình quần thể

- Xác định kích thước mẫn cần thiết để ước lượng trung bình quần thể

2. Khoảng ước lượng cho $\mu$

- Confidence interval : $\bar{X} - E < \mu < \bar{X} + E$ (GioiHanTrungTam)

  + $\bar{X}$ là trung bình của tập mẫu có kích thước $n$
  + $E$ : biên độ lỗi

- Confidence level (level of significance) : là xác suất $1 - \alpha$ mà khoảng tin cậy thực sự chứa tham số quân thể , giả định rằng quá trình ước lượng được lặp đi lặp lại một số lượng lớn lần : $(\alpha = 0.1 ; \alpha = 0.05 ; \alpha = 0.01)$ :

  + Diễn giải : giả sử $\alpha = 0.05%$ ; $2000$ sample có $n = 1007$  , với mỗi lần xây dựng 1 confidence interal thì có 95% khoảng chứa $\mu$ $(\sim 1900$ khoảng)

- $P(\bar{X} - E < \mu < \bar{X} + E) = 1 - \alpha$

- Chọn phân phối :

  - $\sigma$ đã biết or $n > 30$, sử dụng phân phối chuẩn : $\bar{x} - z_{\alpha / 2} * \frac{\sigma}{\sqrt{n}} < \mu < \bar{x} + z_{\alpha / 2} * \frac{\sigma}{\sqrt{n}}$

  - $\sigma$ chưa biết or $n < 30$, sử dụng phân phối $t$ : $\bar{x} - t_{\alpha / 2} * \frac{s}{\sqrt{n}} < \mu < \bar{x} + t_{\alpha / 2} * \frac{s}{\sqrt{n}}$


### **Phân phối t**

$$t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}$$

- Số bậc tự do (df) của tập dữ liệu mẫu là số các giá trị mẫu có thể thay đổi sau khi có rằng buộc cụ thể trên tất cả các giá trị dữ liệu
$$df = n - 1$$

- Tính chất của phân phối $t$ :

  + Phân phối $t$ khác nhau đối với các cỡ mẫu khác nhau
  + Phân phối t có hình dạng chuông đối xứng chung giống như phân phối chuẩn nhưng nó phản ánh sự biến thiên lớn hơn được mong đợi với các mẫu nhỏ
  + Phân phối t có giá trị trung bình là $t = 0$
  + Độ lệch chuẩn của phân phối t thay đổi theo cỡ mẫu và lớn hơn 1
  + khi cỡ mẫu n lớn hơn , phân phối t sẽ gần hơn với phân phối chuẩn

- Tìm kích thước mẫu để ước lượng trung bình quần thể
  + Khi $\sigma$ xác định$$
n = [\frac{z_{a/2}.\sigma}{E}]^{2}
$$

  + khi $\sigma$ không có :
    + $\sigma \approx range / 4$
    + Sử dụng tạm $s$ của mẫu . Giá trị ươc tính sau đó sẽ được cải thiện khi thu được nhiều dữ liệu mẫu hơn và kích thước mẫu có thể được tinh chỉnh cho phù hợp
    + Ước lượng bằng cách sử dụng kết quả của một số nguyên cứu trước đó


### **Ước lượng tỉ lệ quần thể**

- Tỉ lệ quần thể của mẫu $\bar{p} = \frac{x}{n}$
  - x là số phần tử thỏa mãn điều kiện
  - n là tổng số mẫu
- Tỉ lệ mẫu $\bar{p}$ là ước lượng điểm tốt nhất của ước lượng tỉ lệ quần thể $p$

- $\sigma_{\bar{p}} = \sqrt{\frac{p(1-p)}{n}}$

- $\bar{p} - z_{a/2} * \sqrt{\frac{p(1-p)}{n}} \le p \le \bar{p} + z_{a/2} * \sqrt{\frac{p(1-p)}{n}}$

- $n = \frac{z_{a/2}^2 * p(1-p)}{E^2}$

## **Hypothesis Testing**

- Mục tiêu chính của chương này là phát triển khả năng tiến hành kiểm định giả thuyết cho các phát biểu về một trung bình của quần thể $\mu$ , tỷ lệ quần thể $p$

### **Các khái niệm cơ bản**

- Giả thuyết là 1 phát biểu về một thuộc tính của một quần thể . Một kiểm định giải thuyết thông kê là một thủ tục để kiểm định một phát biểu về một thuộc tính của một quần thể

- Giả thuyết null (ký hiệu $H_{0}$) là một phát biểu rằng giá trị của một tham số quần thể bằng với một số giá trị được phát biểu (có thể chứa $\le , \ge$) . Chúng ta kiểm định giả thuyết null trực tiếp theo nghĩa là chúng ta **giả định nó là đúng** và đi đến một kết luận hoặc **bác bỏ $H_{0}$** hoặc **không bác bỏ $H_{0}$**

- Giả thuyết thay thế (ký hiệu là $H_{1}$ hoặc $H_{A}$) là phát biểu rằng tham số có giá trị nào đó khác với giả thuyết null (chỉ sử dụng các ký hiệu như $< , > , \neq$)

- Lưu ý về việc tạo giả thuyết : nếu bạn đang tiến hành một nghiên cứu và muốn sử dụng một kiểm định giả thuyết để hỗ trọ cho phát biểu mà bạn đưa ra , **phát biểu phải được viết để nó trở thành giả thuyết thay thế**

  - VD : hiệu quả nhiên liệu 24 dặm cho mỗi gallon , tạo hệ thống mới để tăng giá trị trung bình về số dặm trên mỗi gallon . Trong trường hợp này , giả thuyết nghiên cứu là 'hệ thống phun nhiên liệu mới này sẽ cung cấp một giá trị trung bình dặm trên mỗi gallon vượt 24' có nghĩa là $\mu > 24$

    - giả thuyết null : $H_{0} : \mu \le 24$
    - giả thuyết thay thế : $H_{1} : \mu > 24$

    - Nếu kết quả mẫu cho thấy $H_{0}$ không thể bị bác bỏ , các nhà nguyên cứu không thể kết luận hệ thống mới tốt hơn
    - Nếu kết quả mẫu cho thấy $H_{0}$ có thể bác bỏ , các nhà nguyên cứu có thể đưa ra suy luận rằng : $H_{1} : \mu > 24$ là đúng (Với kết luận này, các nhà nghiên cứu có bằng chứng về số liệu thống kê để nói rằng hệ thống mới tăng số lượng trung bình dặm trên mỗi galon.)

    - Trong các nghiên cứu như vậy, các giả thuyết null và thay thế nên được biểu diễn để bác bỏ $H_{0}$ và hỗ trợ kết luận cần nghiên cứu. Giả thuyết nghiên cứu do đó nên được thể hiện như là giả thuyết thay thế.  

- Sai lầm loại $I$ là bác bỏ giải thuyết null khi nó đúng ; $\alpha$ được dùng để biểu diễn cho xác suất mắc sai lầm loại I .
  - VD : các nhà nghiên cứu phát biểu rằng hệ thống mới cải thiện dặm trên mỗi gallon , trong thực tế , hệ thống mới không tốt hơn hệ thống hiện tại

- Sai lầm loại $II$ là từ chối bác bỏ giải thuyết null khi nó đúng ; $\beta$ được dùng để biểu diễn cho xác suất mắc sai lầm loại II  
  - VD : các nhà nghiên cứu phát biểu rằng hệ thống mới không tốt hơn hệ thống hiện tại , trong thực tế , hệ thống mới cải thiện hiệu suất dặm mỗi gallon

- Kiểm soát sai lầm loại $I$ và loại $II$
  - Đối với bất kỳ $\alpha$ cố định nào , sự gia tăng cỡ mẫu $n$ sẽ làm giảm $\beta$
  - Đối với bất kỳ cỡ mẫu nào (n) cố định , giảm $\alpha$ sẽ làm tăng $\beta$ . Ngược lại , tăng $\alpha$ sẽ làm giảm $\beta$
  - Để giảm cả $\alpha$ và $\beta$ , hãy tăng kích thước mẫu
- Mức có ý nghĩa (được biểu thị bởi $\alpha$) là xác suất sai lầm loại 'I' xảy ra

## **Kiểm định về giá trị trung bình**

- Khi biết $\sigma$

   Test | Lower Tail Test | Upper Tail Test | Two-Tail Test
   -----| --------------- | --------------- | -------------
   Hypotheses| $H_{0} : \mu \ge \mu_{0}\\ H_{1} : \mu < \mu_{0}$ | $H_{0} : \mu \le \mu_{0}\\ H_{1} : \mu > \mu_{0}$ | $H_{0} : \mu = \mu_{0}\\ H_{1} : \mu \neq \mu_{0}$
   Test Statistic | $z = \frac{x - \mu_{0}}{\frac{\sigma}{\sqrt{n}}}$ | $z = \frac{x - \mu_{0}}{\frac{\sigma}{\sqrt{n}}}$ | $z = \frac{x - \mu_{0}}{\frac{\sigma}{\sqrt{n}}}$
   p-Value| Reject $H_{0}$ if p-value $\le \alpha$ | Reject $H_{0}$ if p-value $\le \alpha$ | Reject $H_{0}$ if p-value $\le \alpha$
   Critical value| Reject $H_{0}$ if z $\le -z_{\alpha}$ | Reject $H_{0}$ if z $\ge z_{\alpha}$ | Reject $H_{0}$ if $z \le -z_{\alpha /2}$ or $z \ge z_{\alpha /2}$

- Khi chưa biết $\sigma$

   Test | Lower Tail Test | Upper Tail Test | Two-Tail Test
   -----| --------------- | --------------- | -------------
   Hypotheses| $H_{0} : \mu \ge \mu_{0}\\ H_{1} : \mu < \mu_{0}$ | $H_{0} : \mu \le \mu_{0}\\ H_{1} : \mu > \mu_{0}$ | $H_{0} : \mu = \mu_{0}\\ H_{1} : \mu \neq \mu_{0}$
   Test Statistic | $t = \frac{x - \mu_{0}}{\frac{s}{\sqrt{n}}}$ | $t = \frac{x - \mu_{0}}{\frac{s}{\sqrt{n}}}$ | $t = \frac{x - \mu_{0}}{\frac{s}{\sqrt{n}}}$
   p-Value| Reject $H_{0}$ if p-value $\le \alpha$ | Reject $H_{0}$ if p-value $\le \alpha$ | Reject $H_{0}$ if p-value $\le \alpha$
   Critical value| Reject $H_{0}$ if t $\le -t_{\alpha}$ | Reject $H_{0}$ if t $\ge t_{\alpha}$ | Reject $H_{0}$ if $t \le -t_{\alpha /2}$ or $t \ge t_{\alpha /2}$
